# Aggregation in tensors

## Get The Minimum

```python
tensor_aggregation=tf.Variable(np.random.randint(1,100,size=(2,50)))
tensor_aggregation
```

```python
tf.reduce_min(tensor_aggregation)
```

## Get The Maximum

```python
tf.reduce_max(tensor_aggregation)
```

## Get The Mean

```python
tf.reduce_mean(tensor_aggregation)
```

## Get The sum

```python
tf.reduce_sum(tensor_aggregation)
```

## Get The absolute

```python
d=tf.constant([-7,-10,2])
d
```

```python
tf.abs(d)
```

## Get The Variance

```python
tf.math.reduce_variance(tf.cast(tensor_aggregation,dtype=tf.float32),axis=0)
```

## Get the SD

```python
sd = tf.math.reduce_std(tf.cast(tensor_aggregation,dtype=tf.float32),axis=0)
sd
```

# find the posation of the max and min

```python
sd[tf.argmax(sd)]
```

```python
sd[tf.argmin(sd)]
```
